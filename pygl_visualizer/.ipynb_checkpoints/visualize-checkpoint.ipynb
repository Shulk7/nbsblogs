```python
win = pyglet.window.Window()

# override the draw method
@win.event
def on_draw():
    # clear all colors from screen
    glClear(GL_COLOR_BUFFER_BIT)
    
    # draw point
    glBegin(GL_POINTS)
    glVertex3f(150., 175., 0.25)
    glVertex3f(160., 160., 0.25)
    glEnd()
    
    #draw line
    glBegin(GL_LINES)
    glVertex3f(50. ,75., 0.25)
    glVertex3f(25., 25., 0.25)
    glEnd()
     
# run our pyglet app
pyglet.app.run()
```

Making a cube! This is a test of 3D features in pyglet more than anything.

```python
from pyglet.window import key

INCREMENT = 5
CENTER = 75

def c(x):
    return CENTER + x

class cubewin(pyglet.window.Window):
    
    xRotation = yRotation = 30
    
    def __init__(self, width, height, title=''):
        super().__init__(width, height, title, resizable=True)
        glEnable(GL_DEPTH_TEST)
        
    def on_draw(self):
        self.clear()
        
        glPushMatrix()

        glRotatef(self.xRotation, 1, 0, 0)
        glRotatef(self.yRotation, 0, 1, 0)

        glBegin(GL_QUADS)
        
        # face 1 
        glColor3ub(255, 255, 255) # white point
        glVertex3f(c(50),c(50),c(50))
        glColor3ub(255, 255, 0)  # yellow point
        glVertex3f(c(50),c(-50),c(50))
        glColor3ub(255, 0, 0) # red side
        glVertex3f(c(-50),c(-50),c(50))
        glVertex3f(c(-50),c(50),c(50))
        
        # face 2 
        glColor3ub(255, 0, 0) # red points
        glVertex3f(c(-50),c(50),c(50))
        glVertex3f(c(-50),c(-50),c(50))
        glColor3f(0, 0, 1) # two blue points
        glVertex3f(c(-50),c(-50),c(-50))
        glVertex3f(c(-50),c(50),c(-50))
        
        # face 3 (top)
        glColor3ub(255, 255, 255) # white point
        glVertex3f(c(50),c(50),c(50))
        glColor3ub(255, 0, 0)
        glVertex3f(c(-50),c(50),c(50)) # top red point
        glColor3f(0, 0, 1) # top blue point
        glVertex3f(c(-50),c(50),c(-50))
        glVertex3f(c(50),c(50),c(-50)) # top point
        
        # face4
        glColor3ub(255,255,255) #white point
        glVertex3f(c(50), c(50), c(50))
        glColor3ub(255,255,0) # yellow point
        glVertex3f(c(50),c(-50),c(50))
        #back bottom point
        glColor3ub(124,252,0)
        glVertex3f(c(50),c(-50),c(-50))
        #top blue point
        glColor3f(0,0,1)
        glVertex3f(c(50),c(50),c(-50))
        
        #face5 
        # top blue point
        glColor3f(0,0,1)
        glVertex3f(c(50),c(50),c(-50))        
        #back bottom point
        glColor3ub(124,252,0)
        glVertex3f(c(50),c(-50),c(-50))
        glColor3f(0,0,1)
        glVertex3f(c(-50),c(-50),c(-50))
        glVertex3f(c(-50),c(50),c(-50))
        
        # face6 bottom
        glColor3ub(255, 255, 0)#yellow point
        glVertex3f(c(50),c(-50),c(50))
        glColor3ub(124,252,0)#green point
        glVertex3f(c(50), c(-50), c(-50))
        glColor3f(0, 0, 1)#blue
        glVertex3f(c(-50),c(-50),c(-50))
        glColor3ub(255, 0, 0) #red
        glVertex3f(c(-50),c(-50),c(50))
        
        glEnd()
        
        glPopMatrix()
        
    def on_resize(self, width, height):
        # set the Viewport
        glViewport(0, 0, width, height)

        # using Projection mode
        glMatrixMode(GL_PROJECTION)
        glLoadIdentity()

        aspectRatio = width / height
        gluPerspective(35, aspectRatio, 1, 1000)

        glMatrixMode(GL_MODELVIEW)
        glLoadIdentity()
        glTranslatef(25, 25 , -800)
        
    def on_text_motion(self, motion):
        if motion == key.UP:
            self.xRotation -= INCREMENT
        elif motion == key.DOWN:
            self.xRotation += INCREMENT
        elif motion == key.LEFT:
            self.yRotation -= INCREMENT
        elif motion == key.RIGHT:
            self.yRotation += INCREMENT

cubewin(400, 400)
# run our pyglet app
pyglet.app.run()
```

```python
import c3d
import pyglet
import numpy as np
from pyglet.gl import *
from pyglet.window import key
# from sklearn.preprocessing import scale

path = '/Users/yvanscher/Downloads/02_01.c3d'

reader = c3d.Reader(open(path, 'rb'))
for i, points, analog in reader.read_frames():
    pos = points[:,:3]
    

INCREMENT = 5
CENTER = -25

# 0, 1, 2 -> 1,2,0
def c(x):
    temp = x[0]
    temp2 = x[2]
    x[0] = x[1]
    x[2] = temp
    x[1] = temp2
    x[0] = x[0] + CENTER
    x[1] = x[1] + CENTER
    return x

class cubewin(pyglet.window.Window):
    xRotation = yRotation = 0
    
    def __init__(self, width, height, title=''):
        super().__init__(width, height, title, resizable=True)
        glEnable(GL_DEPTH_TEST)
        
    def on_draw(self):
        self.clear()
        glPushMatrix()

        glRotatef(self.xRotation, 1, 0, 0)
        glRotatef(self.yRotation, 0, 1, 0)

        glBegin(GL_POINTS)
        for i in range(pos.shape[0]):
            glVertex3f(*c(pos[i]/100))
        glEnd()  
        
        glBegin(GL_POINTS)
        glVertex3f(-25,-25,0)
        glEnd()
        glPopMatrix()
                
    def on_resize(self, width, height):
        glMatrixMode(GL_PROJECTION)
        glLoadIdentity()

        aspectRatio = width / height
        gluPerspective(35, aspectRatio, 1, 1000)

        glMatrixMode(GL_MODELVIEW)
        glLoadIdentity()
        glTranslatef(25, 25 , -50)
        
    def on_text_motion(self, motion):
        if motion == key.UP:
            self.xRotation -= INCREMENT
        elif motion == key.DOWN:
            self.xRotation += INCREMENT
        elif motion == key.LEFT:
            self.yRotation -= INCREMENT
        elif motion == key.RIGHT:
            self.yRotation += INCREMENT

cubewin(400, 400)
# run our pyglet app
pyglet.app.run()
```
