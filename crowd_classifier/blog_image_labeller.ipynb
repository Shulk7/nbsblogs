```python
def show_n(select_imgs, title, n=4):
    '''
    Displays n photos
    '''
    _ , ax = plt.subplots(1, n, sharex='col', sharey='row', figsize=(24, 6))
    plt.suptitle(title, size=20)
    for i, img in enumerate(select_imgs):
        image = Image.open(img, 'r')
        image = image.resize((224,224))
        ax[i].imshow(image)
        img_location = img.replace('/User/yvanscher/Downloads', '')
        ax[i].annotate(img_location,
                      (0,0), (0, -32), xycoords='axes fraction', 
                       textcoords='offset points', va='top')
        ax[i].axis('off')

def chunker(seq, size):
    return (seq[pos:pos + size] for pos in range(0, len(seq), size))

def read_imgs(path):
    return 
```
