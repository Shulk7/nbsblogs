# Question 1 Solve AX = B

solve for $x, y$ given:

$3x - 2y = -2$

$-3x+7y=30$

$A . X = B$

$X = A^{-1} . B$

purpose: show that you understand the fundamentals of matrices, matrix multiplication, and mathematical optimization central to almost all machine learning algorithms.

we rewrite this problem with matrices, multiplying both sides by the inverse of $A$:

$
\begin{bmatrix}
3 & -2 \\ 
-3 & 7
\end{bmatrix} 
\begin{bmatrix}x \\ y\end{bmatrix}
=
\begin{bmatrix}
-2 \\ 
30
\end{bmatrix}
$

$
\begin{bmatrix}
3 & -2 \\ 
-3 & 7
\end{bmatrix}^{-1}
\begin{bmatrix}
3 & -2 \\ 
-3 & 7
\end{bmatrix} 
\begin{bmatrix}x \\ y\end{bmatrix}
=
\begin{bmatrix}
3 & -2 \\ 
-3 & 7
\end{bmatrix}^{-1}
\begin{bmatrix}
-2 \\ 
30
\end{bmatrix}
$

$
\begin{bmatrix}x \\ y\end{bmatrix}
=
\begin{bmatrix}
3 & -2 \\

\end{bmatrix}^{-1} 
\begin{bmatrix}
-2 \\

\end{bmatrix}
$

use the [determinant method](https://www.mathsisfun.com/algebra/matrix-inverse.html) to get $A^{-1}$:

$A^{-1} = 1/(3*7 - {-2}*{-3}) * 
\begin{bmatrix}

$A^{-1} = 1/15 *

# Question 2 Eigenvectors

what is an eigen value? what is an eigenvector? 

an eigenvalue $\lambda$ and an eigenvector $\bar{x}$ of matrix $A$ has the following relationship:

$A\bar{x} = \lambda\bar{x}$

in other words the original matrix can be multiplied with its eigenvector to produce an output. this output will be the same as multiplying the eigenvalue (some scalar constant) with the eigenvector. it's a way to represent a matrix as constant (if you know the eigenvector for that matrix). Using Eigenvalues and Eigenvectors to represent a matrix is called Eigendecomposition (because you are decomposing the matrix into 2 parts).

# Question 3 Solve for the Eigenvectors/Eigenvalues

solve for the eigenvectors and eigenvalues of the matrix:

$
A = \begin{bmatrix}
2 & 7 \\ 
-1 & -6
\end{bmatrix}
$

purpose: show that you understand the fundamental matrix math underlying machine learning.

1 - multiply a 2x2 [identity matrix](https://en.wikipedia.org/wiki/Identity_matrix) by $\lambda$:

$\begin{bmatrix}
\lambda & 0 \\ 
0 & \lambda
\end{bmatrix}$

2 - subtract the outcome of 1 from A

$\begin{bmatrix}
2 & 7 \\ 
-1 & -6
\end{bmatrix}
-
\begin{bmatrix}
\lambda & 0 \\ 
0 & \lambda
\end{bmatrix}
=
\begin{bmatrix}
2-\lambda & 7 \\ 
-1 & -6-\lambda
\end{bmatrix}
$

3 - find the determinant of the outcome of 2

$\det{\begin{bmatrix}
2-\lambda & 7 \\ 
-1 & -6-\lambda
\end{bmatrix}}
=
((2-\lambda)*(-6-\lambda) - (7*-1))
=
\lambda^{2} + 4\lambda -5
$

4 - set the determinant equal to the zero and solve it for values of $\lambda$:

$\lambda^{2} + 4\lambda -5 = 0$

$(\lambda + 5)(\lambda - 1) = 5$

$\lambda=1$

$\lambda=-5$

so now we have ous eigenvalues.

5 - use $A$ and each $\lambda$ to find each eigenvector $\bar{x}$

given $\lambda=1$, $\bar{x} = \begin{bmatrix}x1 \\ x2\end{bmatrix}$:

$A\bar{x} = \lambda\bar{x}$

$
A\bar{x} - \lambda\bar{x} = \begin{bmatrix}
0 \\ 
0
\end{bmatrix}
$

$
(A-\lambda) \bar{x} = \begin{bmatrix}
0 \\ 
0
\end{bmatrix}
$

$\begin{bmatrix}
1 & 7 \\ 
-1 & -7
\end{bmatrix}\begin{bmatrix}x1 \\ x2\end{bmatrix} = 
\begin{bmatrix}
0 \\ 
0
\end{bmatrix}
$

fe can represent this as a system of linear equations ans solve it by plugging in values:

$7x2 + x1 = 0$

$-7x2 - x1 = 0$

so for $\lambda = 1$:

we get the eigenvector as $\bar{x} = \begin{bmatrix}x1 \\ x2\end{bmatrix} = \begin{bmatrix}-7\\ 1\end{bmatrix}$.

for $\lambda = -5$:

$\begin{bmatrix}
7 & 7 \\ 
-1 & -1
\end{bmatrix}\begin{bmatrix}x1 \\ x2\end{bmatrix} = 
\begin{bmatrix}
0 \\ 
0
\end{bmatrix}
$

$7x2 + 7x1 = 0$

$-x2 -x1 = 0$

we get a second eigenvector $\begin{bmatrix}-1 \\ 1\end{bmatrix}$ 

so we have eigenvalues, eigenvectors:

$\lambda = 1, \bar{x} = \begin{bmatrix}-7 \\ 1\end{bmatrix}$

$\lambda = -5, \bar{x} = \begin{bmatrix}-1 \\ 1\end{bmatrix}$
