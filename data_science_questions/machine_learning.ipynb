# optimization

**description**

optimization is the practice of finding a solution which maximizes or minimizes some metric. for example you might choose to minimize a loss function of your model or algorithm. optimization is at the core of machine learning and in a sense it is the mechanism by which the machines learn.

**math:**

often the first type of optimization taught for machine learning is the gradient descent algorithm. below is a graphical represtenation of gradient descent:

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/f/ff/Gradient_descent.svg/512px-Gradient_descent.svg.png"/>

Imagine that the center is a whirlpool or a black hole and we are obsreving it from above, each ring gets steeper as we get closer to the center. your job is to try to ride down into the abyss. 

**example:**

The way gradient descent works is by creating a cost map (in 2, 3, or N dimensions). You create a function that measures your 'cost' as a function of your parameters: $f_c = f_{w1} + f_{w2} + f_{w3} ...$ Then if you change your parameters in this function ($w_1, etc$) the cost would change. After making this function you calculate a 'gradient' for each parameter $w_n$ each gradient the slope (a single number value) of a straight line that tells you how changing that parameter from its current value would affect the cost output by the function. We then always subtract the value of this gradient from the parameter; we go down the slope of the gradient; that is why this method is called gradient descent. Here is an example: 

<img src="https://i.stack.imgur.com/xJgdz.png"/>

Imagine this curve is our cost function with cost on the vertical y axis and our parameter value $w_1$ on the horizonal x axis. You can see that as we adjust our parameter the cost goes up or down. We want to get to the bottom of this function, the lowest point. 

**Let's think of all the cenarios if we start at point E:**

given gradient is -G

adding the gradient:

$E + -G = E - G = F$ (a higher cost value, this is bad)

subtracting the gradient:

$E - -G = E + G = D$ (optimal point, this is good)

**Now let's do the same but assuming we start at point C:**

given gradient is G

adding the gradient:

$C + G = D$ (a higher cost value, bad)

subtracting the gradient:

$C - G = D$ (optimal point, good)

So adding the gradient gives a higher cost value, subtracting it gives a lower one. So we should always subtract our gradient form the current prameter value.

Extra context:

The gradient for a particular parameter $w_1$ com|es from the derivative of your cost function relative to $w_1$ yand we don't really want to get into that. If you ar familiar with calculus the gradient is the slope of the derivative \ trangent line at the point where your parameter currently sits on the cost function.

# learning rates

**description**

a learning rate is the rate at which we accumulate new information into our model. 

**math**

if we have a gradient $G$ we just multiply it by some learning rate $l$.

$E = 5$: our parameter value

$G = -1.25$: our gradient

$l = 0.01$: our learning rate

$E - G*l = 5 - 0.0125 =  4.9875$

This learning rate has moved us along the directon of the gradient but only for a small step.

**example**

You may have noticed above that if we move our parameter too much along the gradient/slope from its current point the value no longer matches our actual cost function. The gradient is only a good estimate of the cost function for very small moves. You can see this below where the straight line function starts to diverge from the actual cost function.

<img src="https://i.stack.imgur.com/xJgdz.png"/>

To deal with this problem we only take a small step along the gradient, we multiply the gradient by some small value (0.01) to reduce the effect of moving along it. This small value is our 'learning rate' sometimes also referred to as a step size.

**description**

the morst basic form of linera regression takes the form $y = b + w_1x_1 + w_2x_2 ...$ where the x variables are the inputs, w are the weights/coefficients for those inputs. and b is a constant term that cna shift the regression up or down along the y axis to get it to fit the data better.

**example**

predicting height from age, gender, and ethnicity. let's solve a regression problem with gradient descent in numpy; see the cell below. The cars dataset is a univariate dataset that measures the stopping distance of the cars with the speed of the cars at the time of craking as an input. Fun fact: these cars are from the 1920s!

The bulk of the gradient descent algorithm for linear regression is in this method, the gradient descent is animated below:

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib import animation, rc
from IPython.display import HTML

# set ffmpeg path to installed path (for anmation)
plt.rcParams['animation.ffmpeg_path'] = '/usr/local/bin/ffmpeg'    

# load the data 
data = pd.read_csv('cars.csv')

m = np.random.normal(loc=10, scale=4, size=1)
b = np.random.normal(loc=1, scale=1, size=1) 
x_data = data['speed']
y_data = data['dist']

def init():
    # start a random line
    line.set_data([], [])
    return (line,) 

# create a graph
fig, ax = plt.subplots()
ax.set_xlim((0,30))
ax.set_ylim((0,130))
_ = ax.scatter(data['speed'], data['dist'])
line, = ax.plot([], [], lw=2)
```

```python
# performs the gradient descent 
# stroing the resulting model at
# each gradient descent step
def calc_regressions(m, b):
    ys = []
    # run 10 iterations of gradient
    # descent
    for i in range(10):
        # create predictions for this
        # m and this b (our parameters)
        y_pred = m * data['speed'] + b
        # calculate the gradients for those parameters
        m_gradient = 2*np.mean(-x_data*(y_data-y_pred))
        b_gradient = 2*np.mean(-(y_data-y_pred))
        # update our parameters with the gradients
        m = m - (m_gradient*0.001)
        b = b - (b_gradient*0.001)
        # store this iteration of gradient descent
        x = np.linspace(0,30)
        ys.append(m * x + b)
    return ys
```

```python
ax.set_title('random reegression lines (random w,b values)')

def animate_rand(i):
    m = np.random.normal(loc=10, scale=4, size=1)
    b = np.random.normal(loc=1, scale=1, size=1) 
    x = np.linspace(0,30)
    y = m * x + b
    line.set_data(x, y)
    return (line,)   

# call the animator for a random line
anim = animation.FuncAnimation(fig, 
                               animate_rand, 
                               init_func=init,
                               frames=10,
                               blit=True)
rc('animation', html='html5')
anim
```

```python
ax.set_title('gradient descent regression values')

ys = calc_regressions(m, b)

# use gradient descent to fit a regression
def animate_fit(i):
    line.set_data(np.linspace(0,30), ys[i])
    return (line,)

# call the animator for our fitted line
anim = animation.FuncAnimation(fig, 
                               animate_fit, 
                               init_func=init,
                               frames=10,
                               blit=True)
rc('animation', html='html5')
anim
```

# regularization

**description**

regularization is a technique used to manage 'high variance' models. you model can be too flexible and overfit the dataset. for example a model that runs through every single point will probably be bad at predicting new points. to smooth out your model you can use regularization to make it fit less closely to the original dataset. models that have been smoothed/regularized tend to predict new data with a lower error than overfitted models.

**math**

regularization is the pracitce of penalizing your model for using too many parameters or a few prameters too heavily. You do this by adding some function of your parameters to your l
oss function. Below I've added two types of regularization to a sum of squared errors loss function.

regression model:

$y = b + w_1x_1 + w_2x_2 ...$

L1 Regularization (aka lasso regression): 

$e = 1/n*\sum(y_i-y_t)^2 + \lambda\sum| w_i | $

lasso regression shrinks small parameters to 0, thus getting rid of them entirely.

L2 Regularization (aka ridge regression):

$e = 1/n*\sum(y_i-y_t)^2 + \lambda\sum{w_i^2} $

ridge regression 

lambda is a hyperparameter that tells the regression 'how much' regularization you want. IF you set it to 0 you get no regularization, all other prositive values increase regularization.

**example**

To regularize linear regeression we can use something called a ridge regression. Ridge regression uses the L2 regularization to regularize our model and often performs better than lasso regression. We will use the R rock dataset which measures the permeability (perm) of petroleum rocks based on, area, perimeter (peri), ans shape of the rocks. This is a regression problem. To thest whether of not ridge regression imporves performance we will split the data into two sections, training and validation datasets. We will train two models, an un-regularized regression and a ridge regression (L2 regularized regression). Then we will test the performance of both these models on the validation data to validate our hypothesis that regularized ridge regression will outperform normal regression on new data.

resources:

https://stats.stackexchange.com/questions/4272/when-to-use-regularization-methods-for-regression

https://medium.com/towards-data-science/l1-and-l2-regularization-methods-ce25e7fc831c

https://stat.ethz.ch/R-manual/R-devel/library/datasets/html/rock.html

```python
import pandas as pd
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import Ridge, LinearRegression

# load the data 
data = pd.read_csv('petroleum_rocks.csv')

x_train, y_train = data[:40][['area','peri','shape']].values.reshape(-1,3), data[:40]['perm'].values.reshape(-1,1)
x_val, y_val = data[40:][['area','peri','shape']].values.reshape(-1,3), data[40:]['perm'].values.reshape(-1,1)

lin_reg = LinearRegression()
# here lambda is called alpha
ridge_reg = Ridge(alpha=0.5)

lin_reg.fit(x_train, y_train), ridge_reg.fit(x_train, y_train)
```

```python
# ridge regression (regularized)
ridge_mse = mean_squared_error(y_val, ridge_reg.predict(x_val))
# linear regression (not regularized)
lin_mse = mean_squared_error(y_val, lin_reg.predict(x_val))
(ridge_mse, lin_mse)
```

ridge_mse: 251611.99 is lower than linear mse: 288385.03 on validation data.
as you can see on the validation data the ridge regression does better. it is regularized and overfits the data less.

# feature selection

**description**

is the process of choosing some subset of features to use in your model and discarding others. some features may be redundant. having too many features can actually make your model worse or increase training time so they need to be removed. There are many types of featurse selection (filter, wrapper, and hybrid-embedded methods). 

filter methods: use statistics like, pearson's correlation, linear discriminant analysis, chi-squared, anova to eliminate variables. For example you may want to pick the top 10 features out of 20 that have the strongest positive or negative correlation to the label/target variable.

wrapper methods: are an iterative process of removing or adding features to make your model better. you can do this a number of ways. one way is to start with all features and iteratively try removing features andother way is to add features. the important concept to take away here is that wrapper methods find a subset, test the perofmance of this subset until either you exhaust all possible subsets or find a model that is good enough for what your task.

trees: You can also use the 'feature importance' feature on certain tree based algorithms to help with selecting the most important features. So for example you can run a random forest and see what features it thinks are important. Always run a few different random forests to see what differences there are in between runs; it may pick some different features each time as most important.

**example:**

We will show an example of using a random forest to select features as this is the fastest to code up and probably what you'll be doing for projects.

other sources:

https://en.wikipedia.org/wiki/Feature_selection

https://www.analyticsvidhya.com/blog/2016/12/introduction-to-feature-selection-methods-with-an-example-or-how-to-select-the-right-variables/

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

# load the data 
data = pd.read_csv('petroleum_rocks.csv')

x_train, y_train = data[:40][['area','peri','shape']].values.reshape(-1,3), data[:40]['perm'].values
x_val, y_val = data[40:][['area','peri','shape']].values.reshape(-1,3), data[40:]['perm'].values

for i in range(5):
    print('forest {}'.format(i))
    rf = RandomForestRegressor()
    rf.fit(x_train, y_train)
    for name, feature_importance in zip(data.columns[:-1], rf.feature_importances_):
        print('name: "{}" importance: {}'.format(name, feature_importance))
```

"area" and "peri" seem to have high numbers (out of 1.0), and "shape" is always 0.1 or below. So it looks like "area" and "peri" (perimeter) are consistently the more important features and shape is less important every time we make a random forest. Let's try to make one regression, called `lin_reg`, with all features ("area" "peri" "shape") and another, called `lin_reg_reduce` with only the two most important features ("area" "peri").

```python
from sklearn.linear_model import LinearRegression

# get two datasets one that has all three vraiables, one that has 
# the two important ones called _reduce and one that has all 3 features
x_train, x_train_reduce, y_train = data[:40][['area','peri','shape']].values.reshape(-1,3), data[:40][['area','peri']].values.reshape(-1,2), data[:40]['perm'].values.reshape(-1,1)
x_val, x_val_reduce, y_val = data[40:][['area','peri','shape']].values.reshape(-1,3), data[40:][['area','peri']].values.reshape(-1,2), data[40:]['perm'].values.reshape(-1,1)

# create 2 linear regression models
# one for each datasets
lin_reg = LinearRegression()
lin_reg_reduce = LinearRegression()

# fit those models
lin_reg.fit(x_train, y_train), lin_reg_reduce.fit(x_train_reduce, y_train)

# see which one performs better on validation data
lin_mse = mean_squared_error(y_val, lin_reg.predict(x_val))
lin_reduce_mse = mean_squared_error(y_val, lin_reg_reduce.predict(x_val_reduce))
(lin_mse, lin_reduce_mse)
```

As you can see by selecting the important features ("area", "peri") and only using those we get a much lower score than just using all the features. lin_reduce_mse is 241601 versus lin_mse which is much higher at 288385.

# feature engineering

**description**

is a process of turning your raw data into a structure that a machine learning algorithm can learn or process (or do those things more easily). some algorithms, like a neural net are themselves a form of feature engineering (the neurons being latent features). feature engineering can include the construction of new features or the transformation of existing features (whitening/normalization).

**math:**

rescale formula:

$x_{scaled} = \frac{x - x_{min}}{x_{max}-x_{min}}$

**example:**

feature engineering can be as simple as creating new features by passing you raw features into functions, for example you could pass your column into a $x^2$ function and get the squared value and pass that into your model. usually randomly passing your features through functions doesnt help much but it can sometimes. feature engineering is probably used most often to 'normalize' data. data is normalized so that all the data in on the same scale. this can sometimes help algorithms make predictions better because it can assign weights more fairly to each feature. below we will go through an example of normalizing data and why it helps.

other sources:

https://en.wikipedia.org/wiki/Feature_engineering

https://medium.com/data-design/what-about-feature-engineering-methods-using-simple-techniques-578ea8c5622d

https://stats.stackexchange.com/questions/111467/is-it-necessary-to-scale-the-target-value-in-addition-to-scaling-features-for-re

```python
data.head()
```

here is our petroleum rock data right above, as you can see area and peri are both quite large in the 1000s but shape , however,  is very small. the fact that removing shape in the last section made the model better should make you suspect that normalization may be helpful. The model may have struggled to use the shape feature because it was different from the other two in a numerical sense. Let's get all these variables on the same scale (0-1.0) and see if performance of a linear regression improves.

```python
# define a rescale function
def rescale_nparray(arr):
    maxval = arr.max()
    minval = arr.min()
    return (arr - minval) / (maxval - minval)

# load the data 
data = pd.read_csv('petroleum_rocks.csv')

data[['area', 'peri', 'shape']] = rescale_nparray(data[['area', 'peri', 'shape']])
data.head()
```

You can see that we have rescaled all our input feature to be between 0-1. we do not usually ned to resclae the target variable. The reasons for this is that it doesn't affect the optimization process. There are some secnarios where sfeature engineering on labels is helpful (label smoothing for GANs) but usually it's not necessary.

```python
from sklearn.linear_model import LinearRegression

# get two datasets one that has all three vraiables, one that has 
# the two important ones called _reduce and one that has all 3 features
x_train, y_train = data[:40][['area','peri','shape']].values.reshape(-1,3), data[:40]['perm'].values.reshape(-1,1)
x_val, y_val = data[40:][['area','peri','shape']].values.reshape(-1,3), data[40:]['perm'].values.reshape(-1,1)

# rescale data for be 0 - 1.0
x_train_re = rescale_nparray(x_train)
x_val_re = rescale_nparray(x_val)

# create 2 linear regression models
# one for raw inputs, one for normalized
# inputs
lin_reg = LinearRegression()
lin_reg_norm = LinearRegression()

# fit those models
lin_reg.fit(x_train, y_train), lin_reg_norm.fit(x_train_re, y_train)

# see which one performs better on validation data
# notice that the validation data also needs to be
# rescaled for our regression model to work right
lin_mse = mean_squared_error(y_val, lin_reg.predict(x_val))
lin_norm_mse = mean_squared_error(y_val, lin_reg_norm.predict(x_val_re))
(lin_mse, lin_norm_mse)
```

As you can see, rescaling the data for the shape parameter improved performance somewhat, but not as much as removing the feature altogether. There is no concrete rule for feature selection and feature engineering. The best way to get it right for a paticular problem is to try different things and find the combinations of feature engineering and selection that work well.

# model selection/evaluation

**description**

a process of choosing the best model among several candidates. often such experiments revolve around determining whether more complex models perform better and exploring this tradeoff for the problem you are trying to solve. often to choose a model models are tested and evaluated on a set of data for which the labels are knonw (a test set). the evaluatio metric is an art unto itself; usually the simpler (single number) the better.

**example:**

data scientists use all sorts of strategies to compare models they've made. often this involves some element of randomness. you can use something called a cross validation where the data is split into sections called folds. Usually 5 folds are used. Then 5 models are trained, each model gets to see 4 of the folds and uses the 5th to evaluate itself. You can use it to compare different models (like a tree model versus a linear regression model) or you can use it to compare the same type but with different hyperparameters. Below I'll do a 5 fold cross validation to show how to pick a model. 


other sources:

https://en.wikipedia.org/wiki/Model_selection

http://scikit-learn.org/stable/modules/model_evaluation.html

```python
import pandas as pd

# load the data 
data = pd.read_csv('petroleum_rocks.csv')

# split the 48 datapoints into roughly equal folds
folds = (data[:10], data[10:20], data[20:30], data[30:40], data[40:])
for i,shape in enumerate(map(np.shape, folds)):
    print('fold {} shape:'.format(i))
    print(shape)
```

as you can see each fold has about 10 datapoints in it, the last having 8. sklearn has [the kfold module](http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.KFold.html) that do this sort of thing automatically and shuffles the folds but i wanted to show the simplest example.

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

# create two arrays for storing validatio nscores
lr_valid_scores = []
rf_valid_scores = []

for i in range(len(folds)):
    # create two models to evaluate
    lr = LinearRegression()
    rf = RandomForestRegressor()

    # get the validation fold, 1 fold
    x_val = folds[i][['area','peri','shape']]
    y_val = folds[i]['perm']
     
    # get the training folds, 4 folds
    other_folds = pd.concat(folds[0:i] + folds[i+1:])

    # get the other folds
    x_train = other_folds[['area','peri','shape']]
    y_train = other_folds['perm']
    
    # fit the models.
    lr.fit(x_train, y_train)
    rf.fit(x_train, y_train)
    
    # calculate the validatio nscores for these models
    lr_mse = mean_squared_error(y_val, lr.predict(x_val))
    rf_mse = mean_squared_error(y_val, rf.predict(x_val))
    
    # store those validation scores
    lr_valid_scores.append(lr_mse)
    rf_valid_scores.append(rf_mse)

# get the average validation score for lr and rf models
np.mean(lr_valid_scores), np.mean(rf_valid_scores)
```

As you can see the linear regression had a lower average loss value than the random forest validating across 5 folds, so we should use the lr model.
